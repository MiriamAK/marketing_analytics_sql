```python
#standard viz library imports
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#Set the directory in the data folder of the repo
%cd ../01_data

ltv_df = pd.read_csv('LTV_calc.csv')
```

```python
ltv_df.columns
```

```python
#calculate the LTV
avg_ltv = np.average(ltv_df['total_purchases_per_month']) * np.average(ltv_df['purchases_per_month']) * np.average(ltv_df['customer_duration (months)'])
```

```python
avg_ltv
```

```python
# See the association between purchases per month and total amount spend per month
plt.scatter('total_purchases_per_month', 'purchases_per_month', data=ltv_df);
plt.title('Total Amount Spent/Mo vs. Number of purchases/mo');
plt.xlabel('Total Amount Spent');
plt.ylabel('Number of purchases/mo');
```
